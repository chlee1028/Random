### Import library

````python
from hdfs3 import HDFileSystem

# Connect to HDFS 
hdfs = HDFileSystem(host='localhost', port=8020)
````

````python
# List files in '/' directory
hdfs.ls('/')
````

````python
# Create a text file
! echo 'hello world!' > hello.txt
````

````python
# Make a directory named test
hdfs.mkdir('/test')
````

````python
# Upload file into test folder
hdfs.put('hello.txt','/test/hello.txt')
````

````python
hdfs.ls('/test')
````

````python
# Read from file
with hdfs.open('/test/hello.txt') as f:
    print(f.read())
````

````python
# Delete the entire directory
hdfs.rm('/test')
````
